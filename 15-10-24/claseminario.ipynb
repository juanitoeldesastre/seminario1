```python
#EJERCICIO 1
#ANALIZAR LOS DATOS DE INFLACION MENSUAL DE PERU DURANTE EL ULTIMO AÑO
#CREAR UNA GRAFICA DE LINEAS PARA VISUALIZAR EL AUMENTO O DISMINUCION
#DE LA INFLACION

import matplotlib.pyplot as plt

#generar data para analisis
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
inflacion = [1.2,1.5,2.0,2.3,2.5,2.7,2.8,2.6,2.1,2.7,2.2,2.0]

#crear grafica
plt.figure(figsize=(10, 6))
plt.plot(meses, inflacion, marker='o', color='red', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Inflacion %', fontsize=14)
plt.title('Inflacion mensual de Peru', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#EJERCICIO 3
#DISTRIBUCION DE CASOS DE COVID-19 POR REGIONES

import seaborn as sns
import matplotlib.pyplot as plt

#crear los datos
regiones = [
    'Lima','Cusco','Arequipa', 'Tacna','Piura','Puno'
    ]
casos = [25000,12000,35000,1000,10000,5800]

#crear la grafica
sns.barplot(x=regiones, y=casos, palette='viridis')
plt.xlabel('Regiones')
plt.ylabel('Casos')
plt.title('Distribucion de casos de COVID-19 por regiones')
plt.show()
```

```python
#VERSION 2 DEL EJERCICIO 3
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#generar los datos aleatorios pa los casos de COVID-19
regiones = [
    'Lima','Cusco','Arequipa', 'Tacna','Piura','Puno'
    ]
casos = np.random.randint(1000,10000,6)

#crear la grafica
sns.barplot(x=regiones, y=casos, palette='viridis')
plt.xlabel('Regiones')
plt.ylabel('Casos')
plt.title('Distribucion de casos de COVID-19 por regiones')
plt.show()

#crear el archivo csv
df = pd.DataFrame({'Regiones': regiones, 'Casos': casos})
df.to_csv('datos_covid.csv', index=False)

'''#descargar el archivo csv
from google.colab import files
files.download('datos_covid.csv')'''

print(df)
```

```python
#EJERCICIO 04 version 2
#MOSTRA SEGUN GRAFICA EL DESEMPLEO EN PERU

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#generar los datos aleatorios para el desempleo
años = ['2018','2019','2020','2021','2022','2023']
desempleo = np.random.uniform(4,20,6) #desempleo 4% hasta 20%

#crear la grafica
plt.figure(figsize=(10, 6))
plt.plot(años, desempleo, marker='o', color='red', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Años', fontsize=12)
plt.ylabel('Desempleo %', fontsize=12)
plt.title('Desempleo en Peru', fontsize=12)
plt.grid(True)
plt.show()

#guardar archivo csv
df=pd.DataFrame({'Años': años, 'Desempleo': desempleo})
df.to_csv('datos_desempleo.csv', index=False)

'''#descargar el archivo csv
from google.colab import files
files.download('datos_desempleo.csv')'''

print(df)

```

```python
#ejercicio 05
#RELACION ENTRE INGRESOS Y NIVELES EDUCATIVOS ///grafica de dispersion
#NIVEL EDUCATIVO : PRIMARIO, SECUNDARIA , TECNICO , UNIVERSITARIO
#INGRESOS PROMEDIO : 800,1000,1500,1800

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Nivel Educativo': ['Primario', 'Secundaria', 'Técnico', 'Universitario'],
    'Ingresos Promedio': [800, 1000, 1500, 1800]
}

df = pd.DataFrame(data)

#crear grafica en de dispersion en matplot
plt.scatter(df['Nivel Educativo'], df['Ingresos Promedio'])
plt.xlabel('Nivel Educativo')
plt.ylabel('Ingresos Promedio')
plt.title('Relación entre Ingresos y Nivel Educativo')
plt.show()

#crear la grafica de dispersion en seaborn
sns.scatterplot(x='Nivel Educativo', y='Ingresos Promedio', data=df)
plt.xlabel('Nivel Educativo')
plt.ylabel('Ingresos Promedio')
plt.title('Relación entre Ingresos y Nivel Educativo')
plt.show()

#crear grafica normal


```

```python
#EJERCICIO 05 VERSION 2

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#generar datos aleatorios

nivel_educativo = ['Primario', 'Secundaria', 'Técnico', 'Universitario']
ingresos_promedio = np.random.uniform(100,8000,4)

#crear la grafica
df = pd.DataFrame({'Nivel Educativo': nivel_educativo, 'Ingresos Promedio': ingresos_promedio})

sns.scatterplot(x='Nivel Educativo', y='Ingresos Promedio', data=df)
plt.xlabel('Nivel Educativo')
plt.ylabel('Ingresos Promedio')
plt.title('Relación entre Ingresos y Nivel Educativo')
plt.show()

print(df)

#guardar csv
df.to_csv('datos_ingresos.csv', index=False)

'''#descargar el archivo csv
from google.colab import files
files.download('datos_ingresos.csv')'''




```
