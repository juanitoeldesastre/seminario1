```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#DATOS FICTICIOS PARA 10 REGIONES
años=['2019','2020','2021','2022','2023']
regiones=['Tacna','Lima','Cusco','Arequipa','Puno','Piura','Ica','Tumbes','Trujillo','Junin']
#matriculas de senati para las 10 regiones (random)
matricula_tacna = np.random.randint(30000,60000,5)
matricula_lima = np.random.randint(90000,100000,5)
matricula_cusco = np.random.randint(40000,80000,5)
matricula_arequipa = np.random.randint(10000,70000,5)
matricula_puno = np.random.randint(9000,50000,5)
matricula_piura = np.random.randint(20000,90000,5)
matricula_ica = np.random.randint(3000,30000,5)
matricula_tumbes = np.random.randint(1000,10000,5)
matricula_trujillo = np.random.randint(40000,120000,5)
matricula_junin = np.random.randint(20000,80000,5)
#CREAR LA DATAFRAME PARA ORGANIZAR LOS DATOS
df=pd.DataFrame({'Años':años,
                 'Matricula Tacna':matricula_tacna,
                 'Matricula Lima':matricula_lima,
                 'Matricula Cusco':matricula_cusco,
                 'Matricula Arequipa':matricula_arequipa,
                 'Matricula Puno':matricula_puno,
                 'Matricula Piura':matricula_piura,
                 'Matricula Ica':matricula_ica,
                 'Matricula Tumbes':matricula_tumbes,
                 'Matricula Trujillo':matricula_trujillo,
                 'Matricula Junin':matricula_junin
                })
#DESVIACION PORCENTUAL POR CADA REGION
desviacion_porcentuales = {
    'Año':años,
    'Desviacion Tacna %':((matricula_tacna-np.mean(matricula_tacna))/np.mean(matricula_tacna))*100,
    'Desviacion Lima %':((matricula_lima-np.mean(matricula_lima))/np.mean(matricula_lima))*100,
    'Desviacion Cusco %':((matricula_cusco-np.mean(matricula_cusco))/np.mean(matricula_cusco))*100,
    'Desviacion Arequipa %':((matricula_arequipa-np.mean(matricula_arequipa))/np.mean(matricula_arequipa))*100,
    'Desviacion Puno %':((matricula_puno-np.mean(matricula_puno))/np.mean(matricula_puno))*100,
    'Desviacion Piura %':((matricula_piura-np.mean(matricula_piura))/np.mean(matricula_piura))*100,
    'Desviacion Ica %':((matricula_ica-np.mean(matricula_ica))/np.mean(matricula_ica))*100,
    'Desviacion Tumbes %':((matricula_tumbes-np.mean(matricula_tumbes))/np.mean(matricula_tumbes))*100,
    'Desviacion Trujillo %':((matricula_trujillo-np.mean(matricula_trujillo))/np.mean(matricula_trujillo))*100,
    'Desviacion Junin %':((matricula_junin-np.mean(matricula_junin))/np.mean(matricula_junin))*100
}
#creamos la DATAFRAME de las desviaciones porcentuales
df_desviaciones=pd.DataFrame(desviacion_porcentuales)
plt.figure(figsize=(12,8))
width=0.20 # ancho de la barra
x=np.arange(len(años)) #posiciones de las barras
 
#crear las barras para cada region
bars_tacna = plt.bar(x-2*width,df_desviaciones['Desviacion Tacna %'],width,label='TACNA')
bars_lima =plt.bar(x-width,df_desviaciones['Desviacion Lima %'],width,label='LIMA')
bars_cusco =plt.bar(x,df_desviaciones['Desviacion Cusco %'],width,label='CUSCO')
bars_arequipa =plt.bar(x+width,df_desviaciones['Desviacion Arequipa %'],width,label='AREQUIPA')
bars_puno =plt.bar(x+2*width,df_desviaciones['Desviacion Puno %'],width,label='PUNO')
bars_piura =plt.bar(x+3*width,df_desviaciones['Desviacion Piura %'],width,label='PIURA')
bars_ica =plt.bar(x+4*width,df_desviaciones['Desviacion Ica %'],width,label='ICA')
bars_tumbes =plt.bar(x+5*width,df_desviaciones['Desviacion Tumbes %'],width,label='TUMBES')
bars_trujillo =plt.bar(x+6*width,df_desviaciones['Desviacion Trujillo %'],width,label='TRUJILLO')
bars_junin =plt.bar(x+7*width,df_desviaciones['Desviacion Junin %'],width,label='JUNIN')
 
#AGREGAR LOS VALORES A LAS BARRAS
for bars in[bars_tacna,bars_lima,bars_cusco,bars_arequipa,bars_puno,bars_piura,bars_ica,bars_tumbes,bars_trujillo,bars_junin]:
  for bar in bars:
    yval=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,yval,f'{yval:.1f}',ha='center',va='bottom',fontsize=6)
 
'''for i in range(len(años)):
  plt.text(x[i]-2*width,matricula_tacna[i]+500,str(matricula_tacna[i]),ha='center')
  plt.text(x[i]-width,matricula_lima[i]+500,str(matricula_lima[i]),ha='center')
  plt.text(x[i],matricula_cusco[i]+500,str(matricula_cusco[i]),ha='center')
  plt.text(x[i]+width,matricula_arequipa[i]+500,str(matricula_arequipa[i]),ha='center')
  plt.text(x[i]+2*width,matricula_puno[i]+500,str(matricula_puno[i]),ha='center')
  plt.text(x[i]+3*width,matricula_piura[i]+500,str(matricula_piura[i]),ha='center')
  plt.text(x[i]+4*width,matricula_ica[i]+500,str(matricula_ica[i]),ha='center')
  plt.text(x[i]+5*width,matricula_tumbes[i]+500,str(matricula_tumbes[i]),ha='center')
  plt.text(x[i]+6*width,matricula_trujillo[i]+500,str(matricula_trujillo[i]),ha='center')
  plt.text(x[i]+7*width,matricula_junin[i]+500,str(matricula_junin[i]),ha='center')
'''
 
 
#configuracion OJOOOOOOOOO
plt.title('Desviacion porcentual de matricula senati por region 2019-2023')
plt.xlabel('Años')
plt.ylabel('Desviacion %')
plt.xticks(x,años)
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
