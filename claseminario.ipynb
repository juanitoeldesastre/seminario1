## Regresi√≥n Lineal Simple
La **Regresi√≥n Lineal Simple** es un m√©todo estad√≠stico que modela la relaci√≥n entre una variable dependiente (resultado) y una variable independiente (predicci√≥n) mediante una l√≠nea recta. La ecuaci√≥n general para una regresi√≥n lineal simple es:

\[
y = mx + b
\]

donde:
- **üìä y** es la variable dependiente.
- **üìà m** representa la pendiente de la l√≠nea.
- **üìâ x** es la variable independiente.
- **‚öì b** es la intersecci√≥n de la l√≠nea en el eje y.


```python
# Importar las librer√≠as necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: cantidad de ventas (X) y precio del producto (y)
X = np.array([[1], [2], [3], [4], [5]]) # Cantidad de ventas
y = np.array([10, 15, 20, 25, 30]) # Precio del producto

# Crear el modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X, y)

# Hacer predicciones
y_pred = model.predict(X)

# Visualizar la l√≠nea de regresi√≥n
plt.scatter(X, y, color='blue', label='Datos Reales') # Puntos de datos
plt.plot(X, y_pred, color='red', label='L√≠nea de Regresi√≥n') # L√≠nea deregresi√≥n
plt.title('Regresi√≥n Lineal Simple')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Precio del Producto')
plt.legend()
plt.show()

# Coeficientes
print(f'Intercepto: {model.intercept_}')
print(f'Pendiente: {model.coef_[0]}')

```

## Ejercicio 1

Dado un conjunto de datos que relaciona las horas de estudio con las calificaciones de los estudiantes, crea un modelo de regresi√≥n lineal simple y predice la calificaci√≥n para 4.5 horas de estudio.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# data
X = np.array([[1], [1.5], [2], [2.3], [3.2], [4], [5], [7.6], [8], [10]])
y = np.array([2, 3, 5, 7, 8.5, 12, 13.5, 14, 18, 19.8])

# creating the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# predictions
calif_pred = model.predict([[4.5]])
print(f'La calificaci√≥n para 4.5 horas de estudio es: {calif_pred[0]}')

# visualizing the data
plt.scatter(X, y, color='blue', label='Datos Registrados Reales')
plt.plot(X, model.predict(X), color='red', label='L√≠nea de Regresi√≥n')
plt.title('Regresi√≥n Lineal Simple Ejercicio 01')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.legend()
plt.show()

# coefficients
print(f'Intercepto: {model.intercept_}')
print(f'Pendiente: {model.coef_[0]}')
```

## Ejercicio 2

Usa datos de la temperatura y el consumo de helados para construir un modelo de regresi√≥n lineal simple. Predice el consumo de helados a una temperatura de 30 grados.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Crear los datos de temperatura y consumo de helados
X = np.array([[10], [13], [17], [20], [22], [28], [36], [38], [40], [44]])  # Temperatura
y = np.array([500, 800, 1186, 3000, 3652, 3856, 4050, 5168, 6002, 7010])   # Consumo de helados

# Crear el modelo
model = LinearRegression()
model.fit(X, y)

# Hacer predicciones
consumo_pred = model.predict([[30]])
print(f'El consumo de helados a una temperatura de 30 grados es: {consumo_pred[0]}')

# Visualizar la l√≠nea de regresi√≥n
plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X, model.predict(X), color='red', label='Regresi√≥n Lineal')
plt.xlabel('Temperatura')
plt.ylabel('Consumo de helados')
plt.legend()
plt.show()

# Valores de los coeficientes
print(f'Intercepto: {model.intercept_}')
print(f'Pendiente: {model.coef_[0]}')
```

## Ejercicio 3 

Crea un modelo de regresi√≥n lineal simple utilizando el n√∫mero de
seguidores de un influencer y las interacciones (likes, comentarios) en sus
publicaciones. Predice las interacciones para un influencer con 20,000 seguidores.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# crear datos de ejemplo de seguidores (x) e interacciones (y)
x = np.array([[1000000], [3000000], [10000000], [21000000], [33000000], [50000000]])  # seguidores
y = np.array([159753, 183957, 258963, 295741, 389472, 807456])  # interacciones

# modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(x, y)

# Hacer la predicci√≥n
interacciones_pred = model.predict(np.array([[34100000]]))
pred_value = interacciones_pred[0]
print(f'Predicci√≥n de interacciones para 34,100,000 seguidores: {pred_value}')

# visualizar la l√≠nea de regresi√≥n
plt.scatter(x, y, color='blue', label='Datos Reales')  # Datos reales
plt.plot(x, model.predict(x), color='red', label='L√≠nea de Regresi√≥n')  # L√≠nea de regresi√≥n
plt.xlabel('Seguidores')
plt.ylabel('Interacciones')
plt.title('Relaci√≥n entre Seguidores e Interacciones')
plt.legend()
plt.show()

# Valores de los coeficientes
print(f'Intercepto: {model.intercept_}')
print(f'Pendiente: {model.coef_[0]}')
```
